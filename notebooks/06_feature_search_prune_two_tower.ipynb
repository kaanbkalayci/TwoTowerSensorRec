```python
from pathlib import Path
import os
import runpy

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

# Train three pruned candidates for 100 epochs:
# A: no RSSI/rank/share, keep spectral core, refined dB + ratios
# B: no RSSI/rank/share, no spectral core, refined dB + ratios
# C: no RSSI/rank/share, keep spectral core, refined dB only
env_updates = {
    "FEATURE_SEARCH_MODE": "candidate",
    "FEATURE_SEARCH_MAX_EPOCHS": "100",
    "FEATURE_SEARCH_PATIENCE": "100",
    "FEATURE_SEARCH_LOG_EVERY": "5",
}
old_env = {key: os.environ.get(key) for key in env_updates}

try:
    os.environ.update(env_updates)
    candidate_ns = runpy.run_path(
        str(PROJECT_ROOT / "scripts" / "feature_search_prune_two_tower.py"),
        run_name="__main__",
    )
finally:
    for key, old_value in old_env.items():
        if old_value is None:
            os.environ.pop(key, None)
        else:
            os.environ[key] = old_value

candidate_results = candidate_ns["comparison_df"]
candidate_results
```

```python
from datetime import datetime, timezone
from pathlib import Path
import json
import os
import runpy
import shutil

import numpy as np
import torch

PROJECT_ROOT = Path.cwd()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

# Train Candidate D longer and promote the selected checkpoint to bestmodel_v3.
# Feature set: per-sensor geometry + 7 refined dB bands; static min13 action tower.
env_updates = {
    "FEATURE_SEARCH_MODE": "bestmodel_v3",
    "FEATURE_SEARCH_MAX_EPOCHS": "250",
    "FEATURE_SEARCH_PATIENCE": "250",
    "FEATURE_SEARCH_LOG_EVERY": "5",
}
old_env = {key: os.environ.get(key) for key in env_updates}

try:
    os.environ.update(env_updates)
    bestmodel_v3_train_ns = runpy.run_path(
        str(PROJECT_ROOT / "scripts" / "feature_search_prune_two_tower.py"),
        run_name="__main__",
    )
finally:
    for key, old_value in old_env.items():
        if old_value is None:
            os.environ.pop(key, None)
        else:
            os.environ[key] = old_value

source_run_name = "v3_drop_rssi_drop_spectral_db_only_h512_d2_e16"
source_run_dir = PROJECT_ROOT / "experiments" / "bestmodel_v3_candidate_training" / source_run_name
best_dir = PROJECT_ROOT / "experiments" / "static_action_two_tower" / "bestmodel_v3"
best_dir.mkdir(parents=True, exist_ok=True)

source_checkpoint = source_run_dir / "selected_checkpoint.pt"
source_embeddings = source_run_dir / "static_action_embeddings.npz"
source_manifest = source_run_dir / "feature_manifest.json"
source_history = source_run_dir / "history.csv"
source_summary = source_run_dir / "summary.csv"
source_common_eval = source_run_dir / "common_eval.csv"

required = [
    source_checkpoint,
    source_embeddings,
    source_manifest,
    source_history,
    source_summary,
    source_common_eval,
]
missing = [str(path) for path in required if not path.exists()]
if missing:
    raise FileNotFoundError("Missing bestmodel_v3 source artifacts: " + ", ".join(missing))

best_checkpoint = best_dir / "bestmodel_v3_complete_checkpoint.pt"
best_state_dict = best_dir / "model_state_dict.pt"
best_embeddings = best_dir / "static_action_embeddings.npz"
best_history = best_dir / "history.csv"
best_summary = best_dir / "summary.csv"
best_common_eval = best_dir / "common_eval.csv"
best_manifest = best_dir / "source_feature_manifest.json"
best_info = best_dir / "bestmodel_v3_info.json"

shutil.copy2(source_checkpoint, best_checkpoint)
shutil.copy2(source_embeddings, best_embeddings)
shutil.copy2(source_history, best_history)
shutil.copy2(source_summary, best_summary)
shutil.copy2(source_common_eval, best_common_eval)
shutil.copy2(source_manifest, best_manifest)

try:
    ckpt = torch.load(best_checkpoint, map_location="cpu", weights_only=False)
except TypeError:
    ckpt = torch.load(best_checkpoint, map_location="cpu")
torch.save(ckpt["model_state_dict"], best_state_dict)

with open(best_manifest, "r", encoding="utf-8") as f:
    manifest = json.load(f)
emb_npz = np.load(best_embeddings, allow_pickle=True)

info = {
    "alias": "bestmodel_v3",
    "model_family": "static_action_two_tower",
    "source_run_name": source_run_name,
    "source_run_dir": str(source_run_dir),
    "best_dir": str(best_dir),
    "complete_checkpoint": str(best_checkpoint),
    "raw_state_dict": str(best_state_dict),
    "static_action_embeddings": str(best_embeddings),
    "history_csv": str(best_history),
    "summary_csv": str(best_summary),
    "common_eval_csv": str(best_common_eval),
    "source_feature_manifest": str(best_manifest),
    "best_epoch": int(ckpt["best_epoch"]),
    "selection_rule": (
        "avg_regret within 3.0% of best-so-far, then mean_rank, top3, "
        "top1, avg_norm_regret, rmse"
    ),
    "regret_rel_tol": float(ckpt.get("regret_rel_tol", 0.03)),
    "context_dim": int(ckpt["context_dim"]),
    "action_dim": int(ckpt["action_dim"]),
    "embedding_shape": list(emb_npz["action_embeddings"].shape),
    "config": ckpt["config"],
    "metrics": ckpt["metrics"],
    "manifest": manifest,
    "created_utc": datetime.now(timezone.utc).isoformat(),
    "note": (
        "bestmodel_v3 is the compact static-action two-tower selected from Candidate D. "
        "The context tower uses only per-sensor normalized geometry and seven 200 ms "
        "refined acoustic dB band levels; the action tower uses only static min13 subset "
        "features, so all 41 action embeddings can be precomputed."
    ),
}

with open(best_info, "w", encoding="utf-8") as f:
    json.dump(info, f, indent=2)

print("Saved bestmodel_v3")
print("best dir:  ", best_dir)
print("checkpoint:", best_checkpoint)
print("embeddings:", best_embeddings)
print("info:      ", best_info)
print("best epoch:", info["best_epoch"])
print("dims:      ", info["context_dim"], "+", info["action_dim"], "=", info["context_dim"] + info["action_dim"])

bestmodel_v3_train_ns["comparison_df"]
```
